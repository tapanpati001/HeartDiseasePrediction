Importing the Dependencies

```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and processing 

```python
#Load the data 
Heart_data=pd.read_csv('heart_disease_prediction.csv')
```

Spliting the Features and Target

```python
X=Heart_data.drop(columns='heart_disease',axis=1)
y=Heart_data['heart_disease']
```

```python
X
```

```python
y
```

spliting the data into training and test data 

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

Logistic Regression 

```python
model= LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

model evalution

```python
#accuracy on training data 
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,y_train)
```

```python
training_data_accuracy
```

Building a predictive System

```python
input_data=(56,1,3,130,256,1,2,142,1,0.6,2,1,6)
#change the input data as numpy array
input_data_as_numpy_array=np.asarray(input_data)
#reshape the numpy array as we are predicting for only one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
prediction=model.predict(input_data_reshaped)

if (prediction[0] == 'presence'):
    print('The person has Heart Disease')
else:
    print('The person does not have  Heart Disease')

```
